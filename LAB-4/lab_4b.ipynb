```python
from nltk.corpus import wordnet as wn
import random
import nltk
nltk.download('omw-1.4')
```

```python
def get_words_from_wordnet(lang):
    all_words = {}
    for synset in wn.all_synsets():
        lemma_names = synset.lemma_names(lang)
        for lemma in lemma_names:
            if lemma not in all_words:
                all_words[lemma] = wn.synsets(lemma)
            else:
                all_words[lemma] += wn.synsets(lemma)
    
    for term in all_words:
        all_words[term] = set(all_words[term])
    return all_words
```

```python
en_w = get_words_from_wordnet('eng')
```

```python
def get_tt_from_t_synset(synsets, tt_lang):
    lemma_freq = {}

    for syns in synsets:
        lemma_names = syns.lemma_names(tt_lang)
        for lemma in lemma_names:
            if lemma not in lemma_freq:
                lemma_freq[lemma] = 1
            else:
                lemma_freq[lemma] += 1
    
    return sorted(lemma_freq.items(), key=lambda item: item[1], reverse=True)
```

```python
tt_lang = "ita_iwn"
```

```python
tot_tt = 0
summ_rid = 0
for w in en_w:
    tt = get_tt_from_t_synset(en_w[w], tt_lang)
    if len(tt) > 1:
        tot_tt += 1
        len_t_senses = len(set(wn.synsets(w)))
        len_t_tt_senses = len(set(wn.synsets(w)).intersection(set(wn.synsets(tt[0][0], lang=tt_lang))))
        tt_rid = round(100-(len(set(wn.synsets(w)).intersection(set(wn.synsets(tt[0][0], lang=tt_lang))))*100/len(set(wn.synsets(w)))), 2)
        print("Sensi associabili a", w + ":", len_t_senses)
        #calcolati come intersezione tra i synset di t e quelli di tt specificando la lingua di tt
        print("Sensi associabili a", w+"-"+tt[0][0] + ":", len_t_tt_senses)
        print("Riduzione dell'ambiguità del",tt_rid,"%")
        summ_rid += tt_rid
        print("")
    
print("Riduzione dell'ambiguità media:", round(summ_rid/tot_tt, 2), "%")
```
